```python
import numpy as np
import matplotlib.pyplot as plt

# Definir los parámetros
slit_distance = 0.0002  # Distancia entre las rendijas
wavelength = 500e-9     # Longitud de onda de la luz
screen_distance = 1.0   # Distancia a la pantalla
screen_width = 0.05     # Ancho de la pantalla
num_points = 1000       # Número de puntos en la pantalla

# La línea de puntos en la pantalla
y_positions = np.linspace(-screen_width / 2, screen_width / 2, num_points)

# Diferencia de caminos ópticos desde cada rendija
d1 = np.sqrt((slit_distance / 2 - y_positions) ** 2 + screen_distance ** 2)
d2 = np.sqrt((-slit_distance / 2 - y_positions) ** 2 + screen_distance ** 2)
diff_path = d1 - d2

# Diferencia de fase entre las ondas
phase_difference = (2 * np.pi * diff_path) / wavelength

# Intensidad resultante por interferencia
intensity = (np.cos(phase_difference / 2)) ** 2

# Normalizar la intensidad para mejor visualización
intensity /= np.max(intensity)

# Graficar el patrón de interferencia
plt.figure(figsize=(10, 6))
plt.plot(y_positions, intensity, label='Interferencia de ondas')
plt.xlabel('Posición en la pantalla')
plt.ylabel('Intensidad relativa')
plt.title('Simulación del experimento de la doble rendija (Ondas)')
plt.legend()
plt.grid()
plt.show()
```
